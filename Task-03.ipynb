 1.数据的合并

```python
#这里的四张表是将原来的train数据分别按行和列分离成的四份
pflu = pd.read_csv("data/train-left-up.csv")
pfld = pd.read_csv("data/train-left-down.csv")
pfru = pd.read_csv("data/train-right-up.csv")
pfrd = pd.read_csv("data/train-right-down.csv")
```

pandas的连接函数concat()函数：
concat函数是在pandas底下的方法，可以将数据根据不同的轴作简单的融合，用作合并和多组数据合并。 series，dataframe或者是panel构成的序列lsit都是可操作的对象。

pd.concat( [df数据1， df数据2， …… ], axis = 0或1， join = 连接方式， keys = 表明数据来源 )

相关参数含义：

axis： 需要合并链接的轴，0是行，1是列

join：连接的方式 ，默认是inner

```python
#用concat（）合并：
result_up1 = pd.concat([pflu, pfru], axis = 1)
result_up1.head()
```

```python
result_down1 = pd.concat([pfld, pfrd], axis = 1)
result1 = pd.concat([result_up1, result_down1], axis = 0)
result1.tail()
```

DataFrame.join(other, on=None, how='left', sort=False)

how参数的使用（下面的merge相同）：{'left'， 'right'， 'outer'， 'inner'}

left：使用调用帧的索引（如果指定了 on，则使用列）

right：使用他人的索引。

outer：将调用帧的索引（或指定了 on 的列）与 other 的索引合并，然后对其进行排序。词典编纂。

inner：形成调用帧的索引（或指定了 on 的列）与其他索引的交集，保留调用的索引的顺序。

cross：从两个帧创建笛卡尔积，保留左键的顺序。

```python
#使用DataFrame自带的方法join方法和append进行合并：
result_up2 = pflu.join(pfru)
result_down2 = pfld.join(pfrd)
#result = result_up2.append(result_down2)
result2 = pd.concat([result_up2,result_down2])
result2.tail()
```

注：append方法目前还能使用，但在新版pandas中，即将被concat方法替代，所以尽量使用功能更加强大的concat方法

DataFrame.merge(Object, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None）

Object是右表，用于和DataFrame进行合并(这个方法用于合并表，想添加样本还需要用append或者concat方法)

how表示

left_on标签或列表，或类似数组，要在左侧数据帧中联接的列或索引级别名称。也可以是左侧数据帧长度的数组或数组列表。这些数组被视为列。

left_index布尔，默认为假。使用左侧数据帧中的索引作为连接键。如果是 MultiIndex，则其他 DataFrame 中的键数（索引或列数）必须与级别数匹配。

```python
#使用Panads的merge方法进行合并：
result_up3 = pflu.merge(pfru, left_index=True,right_index=True)
result_up3.tail()
```

```python
result_down3 = pfld.merge(pfrd, left_index=True,right_index=True)
result3 = pd.concat([result_up3,result_down3])
result3.tail()
result3.to_csv('result.csv')
```

Groupby机制：Hadley Wickham（许多热门R语言包的作者）创造了一个用于表示分组运算的术语"split-apply-combine"（拆分－应用－合并）。第一个阶段，pandas对象（无论是Series、DataFrame还是其他的）中的数据会根据你所提供的一个或多个键被拆分（split）为多组。拆分操作是在对象的特定轴上执行的。例如，DataFrame可以在其行（axis=0）或列（axis=1）上进行分组。然后，将一个函数应用（apply）到各个分组并产生一个新值。最后，所有这些函数的执行结果会被合并（combine）到最终的结果对象中。结果对象的形式一般取决于数据上所执行的操作。

```python
#计算泰坦尼克号男性与女性的平均票价,下面两种写法都能得到结果
pf.groupby(['Sex'])[[ 'Fare']].mean()
```

```python
#通过对比可以看出，groupby要求传入明确的类作为对象，下面的由于在pf中先使用了Fare，所以groupby需要指明Sex数据在哪
FS = pf.Fare.groupby(pf['Sex']).mean()
```

```python
#统计泰坦尼克号中男女的存活人数
SS = pf.Survived.groupby(pf['Sex']).sum()
```

```python
pf.Survived.groupby(pf['Pclass']).sum()
```

DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=NoDefault.no_default, observed=False, dropna=True)

level：int，level name, or sequence of such, default None

如果轴是多索引（分层），则按特定级别分组。

dropna：bool类型参数，默认为True

如果 True，并且组键包含 NA 值，则将删除 NA 值以及行/列。如果为 False，则 NA 值也将被视为组中的键。

```python
#用agg（）方法自定义组合
pf.groupby('Sex').agg({'Fare': 'mean', 'Pclass': 'count'}).rename(columns={'Fare': 'mean_fare', 'Pclass': 'count_pclass'})
```

DataFrame.agg(func=None, axis=0, *args, **kwargs)

对于agg（）方法，最重要的是传入的参数func，可接受的组合包括：

（1）函数

（2）字符串函数名称

（3）函数和/或函数名称列表，例如[np.sum, 'mean']

（4）轴标签的字典 - >函数，函数名称或此类列表。

```python
#得出不同年龄的总的存活人数，然后找出存活人数最多的年龄段，最后计算存活人数最高的存活率（存活人数/总人数）
SA = pf[['Survived']].groupby(pf['Age']).sum()
```

```python
#添加一列，用于统计各年龄段总人数
pf.insert(loc=2, column='Number', value=1) 
```

```python
NA = pf[['Number']].groupby(pf['Age']).sum()
```

```python
survival_rate = pd.concat([SA, NA], axis = 1)
survival_rate.head(20)
```

```python
survival_rate.insert(loc=2,column='Survival_Rate',value=0)
```

```python
#统计各年龄段的存活率
survival_rate['Survival_Rate'] = survival_rate.apply(lambda x: x['Survived'] / x['Number'], axis=1)
survival_rate.head(10)
```

```python
#存活人数最多的年龄段
survival_rate[survival_rate['Survived'].values == survival_rate['Survived'].max()]
```

```python
#最大存活率
survival_rate_max1 =survival_rate.Survived.max() / survival_rate.Survived.sum()
survival_rate_max1
```

```python
survival_rate_max2 =survival_rate.Survived.max() / pf.Survived.sum()
survival_rate_max2
```

为什么两次计算结果不一样呢？因为groupby在进行分组生成新表survival_rate时，参数dropna设置为默认值True，自动去除了原数据中没有年龄标注的样本，造成了样本缺失，导致了最终的统计偏差。

但是反过来思考一下，因为我们无法获知没有年龄标注的样本都处于什么年龄层次，所以忽略这些样本反而可能是最好的选择，所以在统计和年龄相关的最大存活率时，应采用survival_rate的样本数，从分子和分母上都去除有缺失值的样本，而不是将其直接加在分母上。
